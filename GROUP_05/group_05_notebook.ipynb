# Theory behind Decomposition Potential

## What is decomposition potential?

Decomposition potential is the minimum potential required for continuous electrolysis (Blum and Vinal, 1934), it is also known as decomposition voltage. Decomposition potential does not have a well-defined value due to polarisation, and is obtained experimentally through approximation using graphical methods.

## Factors that affect decomposition potential

1. Nature of the electrolyte
2. Concentration of the electrolyte
3. Nature of the electrodes
4. Nature of the products of the electrolysis
5. Temperature
6. Polarisation

## Polarisation & Overvoltage

### Polarisation

All electrochemical reactions occur in a series of steps at the interface between electrode and electrolyte. Polarisation is the difference between the actual electrode potential where electrolysis begins and the standard reduction potential (E°), as it is based on environmental factors, determination of the value of polarisation is often not accurate. The difference is due to current flow or net charge flow (Stern and Geaby, 1957). There are three types of polarisation:

Activation Polarisation: This occurs due to resistance to electrochemical reaction kinetics that occur in the anode and cathode (Sadhukhan, 2017). One example is the accumulation of gases at the interface between the electrode and the electrolyte which results in an activation energy barrier associated with the slowest, rate-limiting step of an electrolytic process.

Concentration Polarisation: Occurs due to changes in concentration in the electrolyte solution at the interface between the electrode and the electrolyte (Perez, 2016), thus the reaction rate is limited by diffusion in the solution.

Ohmic/Resistance Polarisation: Occurs due to charge transport and the intrinsic electrical resistance in cell components like the electrodes and electrolyte. The electromotive force (e.m.f) decreases and the current increases, and vice versa, due to Ohm’s law (V = IR).
    
    V is the voltage (V)
    I is the current (A)
    R is the resistance (Ω)

All cases lead to a decrease in the reduction potential of the electrodes (Kreka et al., 2018), a decrease in the rate of reaction (Yu, 2022) and the reaction eventually comes to a stop (Electrochemical Cells, 2021). Additionally, the applied electric current gets converted into heat instead of the desired electrochemical work (Thakre et al., 2019).

### Overvoltage

Also known as overpotential, overvoltage is needed in all electrolytic processes as it is the additional driving force required to overcome polarisation for the electrolytic process to proceed (Crow, 1974). Activation overvoltage, concentration overvoltage and ohmic overvoltage refer to the overvoltage required to overcome activation polarisation, concentration polarisation and ohmic polarisation respectively.

Overvoltage may be positive or negative. The sign is determined by the nature of the process: overpotentials for reduction are negative, since a more negative potential is more reducing. Conversely, overpotentials for oxidation are positive, since a more positive potential is more oxidizing (Stuve, 2014).

## Importance of decomposition potential

Scale-up factors for practical systems in high-power applications must take into account factors like current distribution across large-area electrodes, heat management and localized overcharge (Takeda et al., 2014). When the energy difference is bigger, the cost will be much more significant.

## Methods

### Theoretical Calculation

E° = -(ΔG°)/nF
  
      E° is the standard potential  
      ΔG is the standard Gibbs free energy  
      n is the number of electrons transferred in the reaction from the balanced reaction  
      F is Faraday's constant (96 500 C/mol)  
      
However, the theoretical calculation of decomposition potential is inaccurate as it does not account for resistance and concentration effects that require overvoltage for the reaction to take place (Electrolysis I, 2016). Thus, decomposition potential can only be approximated through experiments and the use of the graphical method.

To determine decomposition potential, a current-voltage curve is plotted. The value of the decomposition potential is obtained through the extrapolation of tangents at specific points.

As seen in Figure 1, the curve is sectioned into 4 points along the y-axis. The first and last sections are closely observed in all current-voltage curves, where the first branch of the curve is almost horizontal due to the applied voltage being below the deposition value of the cation. When the applied voltage at the electrode is above the deposition value of the cation, more current flows. This increases the value proportionately to the potential difference in excess of the deposition voltage (Cuthbertson and Waddington, 1936).

The sharp gradient after the first section can be explained by the Butler-Volmer equation and the Tafel equation. The onset of oxidation due to overvoltage can be detected by an exponential current increase and a potential plateau as seen in Figure 2 (Homann et al., 2020). The sharp gradient increase can be derived from the Butler-Volmer equation which approximates the real current density–overpotential relation (Ghassemi et al., 2020) and expresses that current has an exponential dependency on overvoltage (Rolison, 1994). The equation also agrees with the Tafel equation to show the exponential relationship between the current and the activation overpotential, reflecting the change rate of the overpotential as a function of the logarithm of current density (Menictas et al., 2014).

The last section of the curve is close to parallel to the y-axis, representing resistance. This indicates full conductivity of the electrolyte where both cations and anions are taking part in current conductivity (Cuthbertson and Waddington, 1936).

#### Two Tangent Method

One way is to draw two tangents: One is at the first non-zero gradient and another one at the sharpest gradient. The x-intercept  of the two tangents will be the decomposition potential. 

The readings to generate the tangent are obtained below and above the decomposition potential. This is because, up to the decomposition potential the current flowing is either zero or very small despite the noticeable increase in voltage. Once the decomposition potential is exceeded, a sharp upturn in current for small increases in applied voltages is observed. Therefore, the voltage at which decomposition is set in must lie at the intersection of the two tangents (Maron and Prutton, 1965).

#### One Tangent Method

The second method only takes the tangent at the sharpest gradient and extrapolates it to find the x-intercept.

It was seen that the last branch of the curve is approximately parallel to the line. The value obtained by producing this part of the curve backwards to cut the x-axis is the decomposition potential, and should theoretically be equal to the corrected voltage corresponding to the last break in the curve (Cuthbertson and Waddington, 1936).

# Code & Discussion

## Concept

For readability of the code, the first non-zero gradient will be referred to as the flattest gradient.

### One Tangent Method



<center> Figure 3: Flowchart on how the code for the one tangent method is developed

The intersection of the gradient for one tangent method can be calculated using:

m = $\dfrac{y_{2}-y_{1}}{x_{2}-x_{1}}$

Given the intersection coordinates is ($x_{2}$,0),

m = $\dfrac{0-y_{1}}{x_{2}-x_{1}}$

$\dfrac{-y_{1}}{m}$ = $x_{2}-x_{1}$

${x_{1}}$ = $\dfrac{x_{2} -y_{1}}{m}$

### Two Tangent Method



<center> Figure 4: Flowchart on how the code for the two tangents method is developed

The intersection (x,y) is obtained by making use of the gradients of two tangents, the tangent with the steepest gradient and the tangent with the flattest gradient, 

Given that $m_{1}$ at $(x_{1}, y_{1})$ is the steepest gradient, $m_{2}$ at $(x_{1}, y_{1})$ is the flattest gradient, 

$m_{1}$ = $\dfrac{y_-y_{1}}{x-x_{2}}$
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; y = $m_{1}({x-x_{1}})+{y_{1}}$ 

$m_{2}$ = $\dfrac{y_-y_{2}}{x-x_{2}}$ 
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; y = $m_{2}({x-x_{2}})+{y_{2}}$ 

Given that the y values are the same, 

$m_{2}({x-x_{2}})+{y_{2}}$ = $m_{1}({x-x_{1}})+{y_{1}}$

$m_{2}x-m_{2}x_{2}+{y_{2}}$  = $m_{1}x-m_{1}x_{1}+{y_{1}}$

$m_{2}x-m_{1}x$  = $y_{1}-y_{2}+m_{2}x_{2}-m_{1}x_{1}$

$(m_{2}-m_{1})x$ = $y_{1}-y_{2}+m_{2}x_{2}-m_{1}x_{1}$

Make x the subject, 

x = $\dfrac{y_{1}-y_{2}+m_{2}x_{2}-m_{1}x_{1}}{m_{2}-m_{1}}$         


## Codes

For our project, we have two functions that can be used. One_tangent is used for the one tangent method and two_tangent is used for the two tangent method. The tangent used in the one tangent method is the steepest gradient (maximum gradient) while the two tangents used in the two tangent method are the steepest and the flattest gradient (smallest non-zero gradient).

### One Tangent Method

```python
def one_tangent(filename):

    df = pd.read_csv(filename)
    voltage = df.iloc[:, 0]
    current = df.iloc[:, 1]
    no_of_values = len(df.index)

    gradient_list = []

    if no_of_values >= 2:  # ensures that there is enough points to compute for gradient
        for i in range(no_of_values-1):
            rise = current[i+1] - current[i]
            run = voltage[i+1] - voltage[i]
            gradient = rise/run
            gradient_list.append(gradient)  # obtain gradient list

    steep_grad = max(gradient_list)
    
    # obtaining the index for the maximum gradient of the list
    steep_index = gradient_list.index(steep_grad)

    # finding x coordinate of tangent given the index
    steep_tangent_voltage_coord = voltage[steep_index]

    # finding y coordinate of tangent given the index
    steep_current_value_at_point = current[steep_index]

    minvoltage = min(voltage)
    maxvoltage = max(voltage)

    # generating the y values to plot the graph
    steep_currentlow = (minvoltage - steep_tangent_voltage_coord) * steep_grad + steep_current_value_at_point
    steep_currenthigh = (maxvoltage - steep_tangent_voltage_coord) * steep_grad + steep_current_value_at_point

    # calculation of intersection of gradient at x axis
    intersect = steep_tangent_voltage_coord - (steep_current_value_at_point/steep_grad)

    # plots the values with a label based on its x coord at tangent
    plt.plot([minvoltage, maxvoltage], [steep_currentlow, steep_currenthigh],
             label=f"Plot when x = {intersect}")

    print(f"The x-intercept is {intersect} which is the decomposition potential.")

    plt.plot(voltage, current)
    plt.ylabel("Current (mA)")
    plt.xlabel("Voltage (V)")
    plt.title("Current (mA) vs Voltage (V)")
    plt.ylim(bottom=min(current)-5, top=max(current)+5)
    plt.legend()
    plt.show()
```

### Two Tangent Method

```python
def two_tangents(filename):
    
    df = pd.read_csv(filename)
    voltage = df.iloc[:, 0]
    current = df.iloc[:, 1]
    no_of_values = len(df.index)

    gradient_list = []
    
    if no_of_values >= 2:
        for i in range(no_of_values-1):
            rise = current[i+1] - current[i]
            run = voltage[i+1] - voltage[i]
            gradient = rise/run
            gradient_list.append(gradient)

    steep_grad = max(gradient_list) 
    steep_index = gradient_list.index(steep_grad)

    flat_grad = next((grad for grad in iter(gradient_list) if grad > 0), None) # finding the flattest gradient
    flat_index = gradient_list.index(flat_grad)

    minvoltage = min(voltage)
    maxvoltage = max(voltage)

    steep_tangent_voltage_coord = voltage[steep_index]
    steep_current_value_at_point = current[steep_index] 

    # generating the min and max steep tangent values using y = mX + C
    currentlow_steep = (minvoltage - steep_tangent_voltage_coord) * steep_grad + steep_current_value_at_point
    currenthigh_steep = (maxvoltage - steep_tangent_voltage_coord) * steep_grad + steep_current_value_at_point

    flat_tangent_voltage_coord = voltage[flat_index] 
    flat_current_value_at_point = current[flat_index]

    # generating the min and max flat tangent values using y = mX + C
    currentlow_flat = (minvoltage - flat_tangent_voltage_coord) * flat_grad + flat_current_value_at_point
    currenthigh_flat = (maxvoltage - flat_tangent_voltage_coord) * flat_grad + flat_current_value_at_point

    plt.plot([minvoltage, maxvoltage], [currentlow_steep, currenthigh_steep],
             label=f"Plot when voltage = {steep_current_value_at_point}")

    plt.plot([minvoltage, maxvoltage], [currentlow_flat, currenthigh_flat],
             label=f"Plot when voltage = {flat_tangent_voltage_coord}")
    
    # finds the intersection of the two lines
    intersect = ((steep_current_value_at_point-flat_current_value_at_point)+(flat_grad *
                   flat_tangent_voltage_coord)-(steep_grad*steep_tangent_voltage_coord))/ (flat_grad-steep_grad)

    print(f"The x value of the intersection of the two tangents is {intersect} which is the decomposition potential")

    plt.plot(voltage, current)
    plt.ylabel("Current (mA)")
    plt.xlabel("Voltage (V)")
    plt.title("Current (mA) vs Voltage (V)")
    plt.ylim(bottom = min(current)-5, top = max(current)+5)
    plt.legend()
    plt.show()
```
